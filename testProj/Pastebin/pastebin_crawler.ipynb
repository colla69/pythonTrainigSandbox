```python
import mechanicalsoup
from json import dump, load
from apscheduler.schedulers.background import BlockingScheduler
import time
from datetime import datetime
import os

def json_save(data, fname):
    with open(fname, 'w') as fp:
        dump(data, fp)

def json_load(fname):
    with open(fname, 'r') as fp:
        return load(fp)
    
scheduler = BlockingScheduler()
job = None
base_uri = "https://pastebin.com"    
browser = mechanicalsoup.StatefulBrowser()
datapath = "files/import"
def save():
    start = datetime.now()
    # paste bin archive aufmachen
    browser.open(base_uri + "/archive")
    page = browser.get_current_page()
    # paste bin archive aufmachen
    
    #html
    table = page.find("table")
    count = 0
    data = {}
    print("start")    
    rows = table.find_all("tr")
    print("running "+str(len(rows))+" rows")
    
    for tr in rows:    
        if count == 0:
            count += 1
            continue              
        for a in tr.find_all("a"):
            try:
                link = base_uri + a["href"]            
                time.sleep(1.5)
                if not "archive" in link:
                    browser.open(link)
                    print( "opening " + link )
                    data[link] = ""
                    single_page = browser.get_current_page()
                    text = single_page.find("textarea",id="paste_code").text
                    data[link] = text
            except Exception as e:
                print(e)
                continue
    
    for link in data:
        if not "archive" in link:
            filename = datapath+link[21:]+".txt"
            if not os.path.isfile( filename ):
                print( "saving .. " + filename )
                file = open(filename, 'w') 
                file.write(data[link]) 
                file.close()

    #json_save(data, datapath)    
    end = datetime.now()
    print( end - start )
    print("done") 

def start_job():
    global job
    scheduler.add_job(save, 'interval', seconds=150)
    save()
    try:
        scheduler.start()
    except:
        pass

start_job()


```
